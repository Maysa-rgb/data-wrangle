```python
import tweepy
from tweepy import OAuthHandler
import json
from timeit import default_timer as timer

# Query Twitter API for each tweet in the Twitter archive and save JSON in a text file
# These are hidden to comply with Twitter's API terms and conditions
consumer_key = ''
consumer_secret = ''
access_token = ''
access_secret = ''

auth = OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)

api = tweepy.API(auth, wait_on_rate_limit=True)


tweet_ids = twitter_archive.tweet_id.values
len(tweet_ids)

# Query Twitter's API for JSON data for each tweet ID in the Twitter archive
count = 0
fails_dict = {}
start = timer()
# Save each tweet's returned JSON as a new line in a .txt file
with open('tweet_json.txt', 'w') as outfile:
    # This loop will likely take 20-30 minutes to run because of Twitter's rate limit
    for tweet_id in tweet_ids:
        count += 1
        print(str(count) + ": " + str(tweet_id))
        try:
            tweet = api.get_status(tweet_id, tweet_mode = 'extended')
            print("Success")
            json.dump(tweet._json, outfile)
            outfile.write('\n')
        except tweepy.TweepError as e:
            print("Fail")
            fails_dict[tweet_id] = e
            pass
end = timer()
print(end - start)
print(fails_dict)
```

```python
# get favorite count and retweet count into dataframe
df =[]
with open('tweet_json.txt') as file :
    for line in file:
        status = json.loads(line)
        tweet_id = status.get("id")
        favorite = status.get("favorite_count")
        retweet = status.get("retweet_count")
        my_data = {'tweet_id' : tweet_id ,
                  'favorite_count': favorite ,
                  'retweet_count':retweet}
        df.append(my_data)
        count_df = pd.DataFrame(df)
        
```

### Analysis

### Is there specific day of week witnessed higher tweets traffic ?

```python
# truncate timestamp column to days only.
twitter_archive_clean['timestamp'] = twitter_archive_clean['timestamp'].dt.date

# convert it again to datetime type because the preceded code converts it to string
twitter_archive_clean['timestamp'] =pd.to_datetime(twitter_archive_clean['timestamp'])
```

```python
day = twitter_archive_clean.groupby(twitter_archive_clean.timestamp.dt.dayofweek)['tweet_id'].count()

# to convert the numerical index to day name
day.index = ['Mon' ,'Tue','Wed','Thu','Fri','Sat','Sun']
day
```

```python
# histogram
plt.figure(figsize =(8,6))
plt.bar([1,2,3,4,5,6,7] , height = day , tick_label = day.index)
plt.title('Descriptive Tweets per Day ' , fontsize =15)
plt.xlabel('Day' ,fontsize = 15)
plt.ylabel('Tweets' ,fontsize = 15)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11);
```

```python
# make inference 
count =[0,0,0,0,0,0,0]
for i in range(10000):
    sample = twitter_archive_clean.sample(twitter_archive_clean.shape[0] , replace = True)  
    day = sample.groupby(sample.timestamp.dt.dayofweek)['tweet_id'].count()
    count = [count[i] + day[i] for i in range(len(count))]
```

```python
plt.figure(figsize =(8,6))
labels = ['Mon','Tue','Wed','Thu','Fri','Sat','Sun']
plt.bar([1,2,3,4,5,6,7] , height = day , tick_label = labels)
plt.title('Inferential Tweets per Day ' , fontsize =15)
plt.xlabel('Day' ,fontsize = 15)
plt.ylabel('Tweets' ,fontsize = 15)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11);
```

**Monday witnesses the highest tweets traffic and sunday witnesses the lowest**
